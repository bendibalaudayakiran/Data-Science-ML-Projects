```python
#lets divide the dataset into two, one is input and another is output dataset
df_x=df_wine.drop(columns=["diluted wines"])
y=df_wine[["diluted wines"]]
```

```python
#Lets chcek the skewness of the input dataset
print(df_x.skew())
print()
print(len(df_x.skew().loc[df_x.skew()>0.55]))# Number of features having skewness more than 0.55
```

```python
#Lets treat above skewness
import numpy as np
for index in df_x.skew().index:
    if df_x.skew().loc[index] >0.55:
        df_x[index]=np.cbrt(df_x[index])
```

```python
#lets again check the skewness
print(df_x.skew())
print()
print(len(df_x.skew().loc[df_x.skew()>0.55]))# Number of features having skewness more than 0.55
```

```python
#lets create a function for maxr2_score to perform different models

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#lets make use of KNN regressor
#we will use pipeline to bring bring features to common scale
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
knr=KNeighborsRegressor()
pipeline=Pipeline([("ss",StandardScaler()),("knr",KNeighborsRegressor())])
parameters = {"knr__n_neighbors":range(2,101)}
clf = GridSearchCV(pipeline, parameters, cv=5,scoring="r2")
clf.fit(df_x, y)
clf.best_params_
```

```python
pipeline_knr=Pipeline([("ss",StandardScaler()),("knr",KNeighborsRegressor(n_neighbors=13))])
#knr=KNeighborsRegressor(n_neighbors=13)
maxr2_score(pipeline_knr,df_x,y)
```
